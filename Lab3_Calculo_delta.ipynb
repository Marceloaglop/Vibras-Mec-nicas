# Codigo lab 02

```python
# Codigo de referencia del lab 2
def X_div_delta_st(r_array, xi):
    analitico = 1 / np.sqrt( (1-r_array**2)**2  + (2*xi*r_array)**2 )
    return analitico
#####################################################################
arr_exp = np.linspace(0.6,1.6,1000)
xi_exp = 0.011

#####################################################################
m0 = 28.08/1000 # [kg]
r0 = 45 /1000 # [m]
w_array = 2*np.pi*np.array([3.25, 4, 4.3, 4.5, 5.5, 5.7, 6, 6.75])

k = 2.5 * 10**3 # [N/m]
F0 =  2*m0 * w_array **2 * r0

delta_st = F0/k
X = np.array([0.0041, 0.013, 0.028, 0.061,0.017,0.015,0.013,0.01])

Xdivdelta = X/delta_st

w_natural = 29.651634463541193
w_natural = 29.652

r_array = w_natural /w_array
####################################################################
# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))
# Plotear los datos
plt.plot(arr_exp, X_div_delta_st(r_array=arr_exp,xi=xi_exp), label='X/d', color='b', linewidth=2)
plt.plot(r_array, Xdivdelta, 'o-', markerfacecolor='none', color = "red", markeredgewidth=1.5)
# Etiquetas de los ejes
plt.xlabel(r'$r$', fontsize=14)
plt.ylabel(r'$\frac{X}{\delta_{st}}$', fontsize=14)  # Formato de fracción en LaTeX
# Título del gráfico
# Leyenda
# Grid para mejor visualización
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
# Mostrar el gráfico
plt.show()
```
